```python
X1 = df1.drop("price", axis=1)
y1 = df1["price"]

from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(X1_train, y1_train)
y1_pred = model1.predict(X1_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

print("R²:", r2_score(y1_test, y1_pred))
print("MAE:", mean_absolute_error(y1_test, y1_pred))
print("RMSE:", np.sqrt(mean_squared_error(y1_test, y1_pred)))
```

```python
coeff_df1 = pd.DataFrame(model1.coef_, X1.columns, columns=["Coefficient"])
print(coeff_df1)
```
